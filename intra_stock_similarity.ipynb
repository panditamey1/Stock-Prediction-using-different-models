```python
data = pd.read_csv("RELIANCE-EQ_2018_2023_intra_five_min.csv")
```

```python
data.drop(["V"], axis = 1, inplace = True)
```

```python
data['date'] = data['timestamp'].map(lambda x : x.split(" ")[0])
```

```python
import time
```

```python
reliance_df = pd.DataFrame()
cols = []
flag = True
for name, df in data.groupby("date"):
    print (name)
    df = df[:75]
    print (len(df))
    if len(df) < 75:
      continue
    #print(df.drop(['timestamp','date'], axis = 1).T.head())
    df2 = df.drop(['timestamp','date'], axis = 1)
    df2.index = df2.index + 1
    df_out = df2.stack()
    df_out.index = df_out.index.map('{0[1]}_{0[0]}'.format)

    df2 = df_out.to_frame().T

    #df2.insert(0, 'date', [name])
    df2.insert(0, "date", pd.Series([name], index=[0]))
    if flag:
      cols = df2.columns
      flag = False
    #print(df2)
    df2.columns = cols
    reliance_df = pd.concat([reliance_df,  df2])
    #print(reliance_df)
    #time.sleep(2)
    #break
```

```python
reliance_df.reset_index(drop=True, inplace = True)
```

```python
reliance_df.to_csv("reliance_2018_2023.csv", index = False)
```

```python
def get_ohlc_data(test_point):
  cols = test_point.columns
  print(cols)
  if 'date' in cols:
    cols = cols[1:]
    test_point = test_point[cols]
  test_df_pt = pd.DataFrame()
  for x in range(0, len(cols) -  3,4):
    num = x 

    #print((test_point[cols[(num + 1)]].values[0]))
    #break
    #d2 = pd.DataFrame(data = {"O" : [test_point['O_' + str(num + 1)]],"H" : [test_point['H_' + str(num + 2)]],"L" : [test_point['L_' + str(num + 3)]],"C" : [test_point['C_' + str(num + 4)]]})
    d2 = pd.DataFrame(data = {"O" : [test_point[cols[(num )]].values[0]],"H" : [test_point[cols[(num + 1)]].values[0]],"L" : [test_point[cols[(num + 2)]].values[0]],"C" : [test_point[cols[(num + 3)]].values[0]]})
    #print(d2)
    test_df_pt = pd.concat([test_df_pt, d2 ])
    #print(test_df_pt)
  return test_df_pt 
```
